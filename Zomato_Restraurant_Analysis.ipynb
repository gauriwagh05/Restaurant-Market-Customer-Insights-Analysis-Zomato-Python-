```python
cuisine_cols = [col for col in df.columns if'cuisines_'in col]
```

```python
cuisine_rating = {}
```

```python
for col in cuisine_cols:
    cuisine_rating[col.replace('zomato_project_3.cuisines_', '')] = (
        df[df[col] == 1]['aggregate_rating'].mean()
    )
```

```python
cuisine_rating_df = (
    pd.DataFrame.from_dict(cuisine_rating, orient='index', columns=['avg_rating'])
    .sort_values(by='avg_rating', ascending=False)
    .head(10)
)

cuisine_rating_df
```

```python
import matplotlib.pyplot as plt

# Plot bar chart
cuisine_rating_df.plot(
    kind='bar',
    figsize=(10,5),
    legend=False,
    title='Top 10 Cuisines by Average Rating'
)

plt.ylabel('Average Rating')
plt.xlabel('Cuisine')
plt.xticks(rotation=45)
plt.show()
```
